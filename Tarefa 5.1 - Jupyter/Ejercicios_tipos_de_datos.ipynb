### Exercicio 7: Lista de Diccionarios

Crea unha lista de diccionarios con información de estudantes:

```python
estudantes = [
    {"nome": "Ana", "idade": 20, "nota": 8.5},
    {"nome": "Carlos", "idade": 22, "nota": 7.8},
    {"nome": "María", "idade": 21, "nota": 9.2}
]
```

Realiza estas operacións:

1. Engade un nopvo estudante.
2. Atopa o estudiante con maior nota.
3. Calcula o promedio de idades.
4. Crea unha lista só cos nomes.
5. Ordena os estudantes por nota (descendente).

```python
estudantes.append({"nome": "Juan", "idade": 23, "nota": 6.5})
mellor = max(estudantes, key=lambda x: x["nota"])
media_idade = sum(e["idade"] for e in estudiantes) / len(estudantes)
nomes = [e["nome"] for e in estudiantes]
estudantes.sort(key=lambda x: x["nota"], reverse=True)

print(f"Mellor nota: {mellor['nome']} ({mellor['nota']})")
print(f"Media idades: {media_idade} | Nomes: {nomes}")
```

### Exercicio 8: Conxuntos

A partir destes conxuntos:

```python
conx_a = {1, 2, 3, 4, 5}
conx_b = {4, 5, 6, 7, 8}
```

Realiza estas operacións:

1. Unión de ambos conxuntos
2. Intersección
3. Diferencia (A - B)
4. Diferencia simétrica
5. Verifica si 3 está en A

```python
union = conx_a | conx_b
interseccion = conx_a & conx_b
diferencia = conx_a - conx_b
simetrica = conx_a ^ conx_b
existe_tres = 3 in conx_a

print(f"Unión: {union}\nIntersección: {interseccion}\nDiferencia: {diferencia}")
print(f"Simétrica: {simetrica}\n¿3 está en A?: {existe_tres}")
```

### Exercicio 9: Programa Integrador

Crea un programa que:

1. Pida ao usuario o seu nome, idade e cidade de residencia
2. Almacene esa información nun diccionario
3. Convirta a idade a enteiro
4. Cree unha lista de hobbies (pregunta por 3 polo menos)
5. Amose toda a información formateada de xeito lexible

```python
usuario = {
    "nome": input("Nome: "),
    "idade": int(input("Idade: ")),
    "cidade": input("Cidade: ")
}
hobbies = []
for i in range(3):
    hobbies.append(input(f"Hobby {i+1}: "))
usuario["hobbies"] = hobbies

print("\n--- Información do Usuario ---")
for k, v in usuario.items():
    print(f"{k.capitalize()}: {v}")
```

### Exercicio 10: Combinación de Tipos Complexos

Temos un Sistema de biblioteca:

```python
biblioteca = {
    "libros": [
        {
            "titulo": "Cien años de soledad",
            "autor": "Gabriel García Márquez", 
            "ano": 1967,
            "xeneros": {"Realismo mágico", "Novela"},
            "disponible": True
        },
        {
            "titulo": "1984",
            "autor": "George Orwell",
            "ano": 1949,
            "xeneros": {"Ciencia ficción", "Distopía"},
            "disponible": False
        }
    ],
    "usuarios": {
        "001": {"nome": "Ana", "libros_prestados": ["1984"]},
        "002": {"nome": "Carlos", "libros_prestados": []}
    }
}
```

Realiza estas operacións:

1. Engade un novo libro.
2. Marca un libro como non dispoñible.
3. Atopa todos os libros dun xénero.
4. Libros dispoñibles por autor.
5. Usuario con máis libros prestados.

```python
# 1. Novo libro
biblioteca["libros"].append({
    "titulo": "The Hobbit", "autor": "Tolkien", "ano": 1937,
    "xeneros": {"Fantasía"}, "disponible": True
})
# 2. Marcar non dispoñible
biblioteca["libros"][0]["disponible"] = False
# 3. Libros dun xénero
ciencia_ficcion = [l["titulo"] for l in biblioteca["libros"] if "Ciencia ficción" in l["xeneros"]]
# 4. Dispoñibles por autor
dispo_autor = {l["autor"]: l["titulo"] for l in biblioteca["libros"] if l["disponible"]}
# 5. Usuario con máis libros
mais_prestados = max(biblioteca["usuarios"].values(), key=lambda u: len(u["libros_prestados"]))

print(f"Ciencia Ficción: {ciencia_ficcion}")
print(f"Dispoñibles: {dispo_autor}")
print(f"Usuario con máis préstamos: {mais_prestados['nome']}")
```

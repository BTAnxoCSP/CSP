### Exercicio 1: Operadores Aritméticos

Calcula a área e perímetro dun rectángulo:
- Pide ao usuario base e altura.
- Calcula área.
- Calcula perímetro.
- Utiliza `//` e `%`.
- Amosa os resultados utilizando diferentes formatos de cadeas.

```python
base = float(input("Base: "))
altura = float(input("Altura: "))

area = base * altura
perimetro = 2 * (base + altura)

print(f"Área: {area} | Perímetro: {perimetro}")
print("División entera base//altura: {} | Resto base%altura: {}".format(base // altura, base % altura))
```

### Exercicio 2: Operadores de Asignación

Simula unha conta bancaria:
- Saldo inicial: 1000
- Usa += para depositar 500
- Usa -= para retirar 300
- Usa *= para aplicar interés do 10%
- Usa /= para dividir entre 2 (metade para aforros)
- Amosa saldo despois de cada operación
- Utiliza diferentes formatos de cadeas.

```python
saldo = 1000
print(f"Saldo inicial: {saldo}")
saldo += 500
print(f"Tras depósito: {saldo}")
saldo -= 300
print(f"Tras retiro: {saldo}")
saldo *= 1.10
print(f"Tras interés: {saldo:.2f}")
saldo /= 2

print("Saldo final (aforros): %s" % saldo)
```

### Exercicio 3: Operadores de Comparación

Sistema de verificación de acceso:
- Pide idade e se é socio ou non.

- Verifica se idade <> 0
- Usa todos estes operadores: >, <, >=, <=, ==, !=

### Exercicio 4: Operadores Lóxicos

Determina se un ano é bisesto:
- Divisible entre 4.
- Non divisible entre 100, excepto se é divisible entre 400.
- Non encadenes condicionais senón condicións.
- Exemplo: 2024 (bisesto), 1900 (non bisesto), 2000 (bisesto).

```python
ano = int(input("Ano: "))

bisesto = (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0)

print(f"O ano {ano} é bisesto: {bisesto}")
```

### Exercicio 6: Operadores de Pertenza

Buscador de vogais en texto:
- Pide unha frase ao usuario.
- Usa `in` para verificar se contén cada vogal.
- Usa `not in` para verificar se non contén números.
- Conta cantas vogais diferentes ten.

```python
frase = input("Frase: ").lower()

v_a = "a" in frase
v_e = "e" in frase
v_i = "i" in frase
v_o = "o" in frase
v_u = "u" in frase

sen_numeros = True
for n in "0123456789":
    if n in frase:
        sen_numeros = False
        break

diferentes = sum([v_a, v_e, v_i, v_o, v_u])

print(f"A frase non contén números: {sen_numeros}")
print(f"Número de vogais diferentes: {diferentes}")
```

### Exercicio 9: Combinación de Operadores

Calculadora de IMC (xa a fixemos pero desta volta é máis completa):
- peso / altura²
- Clasifica segundo táboa OMS
- Baixo peso: < 18.5
- Normal: 18.5 - 24.9
- Sobrepeso: 25 - 29.9
- Obesidade: >= 30
- Usa operadores de comparación e lóxicos
- Opcional: walrus para evitar repetir cálculo

```python
peso = float(input("Peso (kg): "))
altura = float(input("Altura (m): "))

if (imc := peso / (altura**2)) < 18.5:
    cat = "Baixo peso"
elif 18.5 <= imc <= 24.9:
    cat = "Normal"
elif 25 <= imc <= 29.9:
    cat = "Sobrepeso"
else:
    cat = "Obesidade"

print(f"IMC: {imc:.2f} | Categoría: {cat}")
```

### Exercicio 10: Analizador de Texto

Analiza un texto introducido polo usuario:
1. Lonxitude.
2. Se contén certas palabras.
3. Conta vogais.
4. Verifica se é palíndromo.
5. Verifica se texto é `None`.
6. Concatena.

### Xogo de Adiviñanza

Xogo onde:
- Número secreto aleatorio (1-100).
- Aportar Pistas ("é maior", "é menor").
- Contador de intentos.
- Utiliza Walrus.
- Preguntar se se quere xogar outra vez.

```python
import random

xogar = "si"
while xogar == "si":
    secreto = random.randint(1, 100)
    intentos = 0
    while (num := int(input("Adiviña (1-100): "))) != secreto:
        intentos += 1
        print("É maior" if num < secreto else "É menor")
    print(f"Gañaches en {intentos + 1} intentos!")
    xogar = input("Xogar outra vez? (si/no): ").lower()
```

### Validador de Email

Valida email con:
- Que conteña unha `@`.
- Que conteña un `.` despois da `@`.
- Lonxitude >= 5.
- Non conteña espazos en branco.
- Dominio pertenza a unha lista de dominios válida.

```python
email = input("Email: ")

dominios = ["gmail.com", "hotmail.com", "iessanclemente.net"]
valido = "@" in email and "." in email.split("@")[-1] and len(email) >= 5 and " " not in email and email.split("@")[-1] in dominios

print(f"Email válido: {valido}")
```
